```python


url='https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=1900-01-01&minlatitude=34.452&maxlatitude=37.44&minlongitude=-91.033&maxlongitude=-79.915&eventtype=earthquake'
```

```python
usgs_response=requests.get(url)
usgs_response
```

```python
usgs_json=usgs_response.json()
```

```python

usgs_json['features']
```

```python
usgs_data = pd.json_normalize(usgs_json,  'features', sep = '_')
```

```python
usgs_data.head(20)
```

```python
usgs_data.tail(60)
```

```python
usgs_data.columns = [col.replace('properties_', '') for col in usgs_data.columns]
usgs_data.info()
```

```python
usgs_data['title'].unique()
```

```python
#usgs_data.columns = usgs_data.columns.str.strip('properties.')
#usgs_data.columns = usgs_data.columns.str.lstrip('\properties.')
```

```python
usgs_data['TN'] = usgs_data['place'].str.contains('Tennessee')
```

```python
usgs_tn= usgs_data[usgs_data['TN']==True]
```

```python
usgs_tn.query('mag>=2.5', inplace=True)
usgs_tn.shape
```

```python
usgs_tn=pd.DataFrame(usgs_tn)
usgs_tn[['longitude', 'latitude', 'depth']] = pd.DataFrame(usgs_tn['geometry_coordinates'].tolist(), index= usgs_tn.index)
usgs_tn.head(60)
```

```python
usgs_tn.tail(60)
```

```python
usgs_tn.info()
```

```python
usgs_tng=usgs_tn[['mag', 'place', 'longitude', 'latitude', 'depth']]
```

```python
geometry = [Point(xy) for xy in zip(usgs_tng['longitude'], usgs_tng['latitude'])]
```

```python
crs = {'init': 'epsg:4326'}
usgs_tn_geo = GeoDataFrame(usgs_tng, crs=crs, geometry=geometry)
```

```python
usgs_tn_geo.head(60)
```

```python
m = folium.Map([35.01, -85.27], zoom_start= 8, tiles='cartodbpositron')
folium.GeoJson(usgs_tn_geo).add_to(m)

m
```

```python
locations = usgs_tng[['latitude', 'longitude']]
coords = locations.values.tolist()
```

```python

from folium.plugins import MarkerCluster

m = folium.Map([35.01, -85.27], zoom_start= 8, tiles='cartodbpositron')

marker_cluster = MarkerCluster().add_to(m)

for point in range(0, len(coords)):
    folium.Marker(location = coords[point], popup= str(usgs_tng['depth'])  +  str(usgs_tng['mag'])).add_to(marker_cluster)

m
```
